```python
def floatToGms(c):
    grados = math.trunc(c/10000)
    
    if grados > 99:
        minutos = str(c)[3:5]
        segundos = str(c)[5:7]
    else:
        minutos = str(c)[2:4]
        segundos = str(c)[4:6]
    
    
    return grados, int(minutos), int(segundos)
```

```python
def floatToDec(c):
    grados = math.trunc(c/10000)
    
    if grados > 99:
        minutos = str(c)[3:5]
        segundos = str(c)[5:7]
    else:
        minutos = str(c)[2:4]
        segundos = str(c)[4:6]
    
    dec = grados + int(minutos)/60 + int(segundos)/3600
    
    return dec
```

```python
latitud_series = localidades["latitud"].apply(floatToDec)
```

```python
latitud_series
```

```python
localidades.head()
```

```python
localidades["latitud_dec"] = latitud_series
```

```python
localidades.head()
```

```python
longitud_series = localidades["longitud"].apply(floatToDec)
```

```python
longitud_series
```

```python
localidades["longitud_dec"] = longitud_series
```
