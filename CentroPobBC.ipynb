### Centro medio de población Baja California

```python
import pandas as pd
import math
import numpy
```

```python
localidades = pd.read_csv("iter_00_cpv2010.csv", encoding="utf-8")
```

```python
localidades.head()
```

```python
localidades.dropna(subset = ["tam_loc"], inplace=True)
```

```python
localidades
```

```python
bajaCalifornia = localidades[localidades["nom_ent"] == "Baja California"]
```

```python
bajaCalifornia
```

```python
def floatToDec(c):
    grados = math.trunc(c/10000)
    
    if grados > 99:
        minutos = str(c)[3:5]
        segundos = str(c)[5:7]
    else:
        minutos = str(c)[2:4]
        segundos = str(c)[4:6]
    
    dec = grados + int(minutos)/60 + int(segundos)/3600
    
    return dec
```

```python
latitud_dec = bajaCalifornia["latitud"].apply(floatToDec)
longitud_dec = bajaCalifornia["longitud"].apply(floatToDec)
```

```python
latitud_dec
```

```python
bajaCalifornia["latitud_dec"] = latitud_dec
```

```python
bajaCalifornia
```

```python
bajaCalifornia["longitud_dec"] = longitud_dec
```

```python
bajaCalifornia
```

```python
poblacionTotal = bajaCalifornia["pobtot"].sum()
```

```python
poblacionTotal
```

```python
pobXLat = bajaCalifornia.pobtot * bajaCalifornia.latitud_dec
```

```python
pobXLat
```

```python
pobXLat.sum()
```

### Latitud del centro medio de población

```python
pobXLat.sum() / poblacionTotal
```

```python
bajaCalifornia[bajaCalifornia['latitud_dec'].between(32.39, 32.4)].sort_values("pobtot", ascending=False)
```

```python
pob_lon_num = bajaCalifornia.pobtot * bajaCalifornia.longitud_dec * numpy.cos(bajaCalifornia.latitud_dec * (math.pi / 180))
```

```python
pob_lon_den = bajaCalifornia.pobtot * numpy.cos(bajaCalifornia.latitud_dec * (math.pi / 180))
```

### Longitud del centro medio de población

```python
pob_lon_num.sum() / pob_lon_den.sum()
```

```python
bajaCalifornia[bajaCalifornia['longitud_dec'].between(116.43, 116.44)].sort_values("pobtot", ascending=False)
```

```python
bajaCalifornia.query('longitud_dec >= 116.4 & longitud_dec <= 116.5 & latitud_dec >= 32.3 & latitud_dec <= 32.6').sort_values("pobtot", ascending=False)
```
